```python
values = {
    "arithoh.sh": [15.16, 15.16, 0.0],
    "fstime.sh": [14.33, 0.30, 3.01],
    "syscall.sh": [4.46, 1.25, 3.21],
    "pipe.sh": [6.50, 0.85, 5.65],
    "spawn.sh": [5.46, 0.21, 4.96]
}
```

```python
import pandas as pd

real = [x[0] for x in list(values.values())]
user = [x[1] for x in list(values.values())]
sys = [x[2] for x in list(values.values())]

index = list(values.keys())
df = pd.DataFrame({'real': real,
                   'user': user,
                   'sys': sys
                   }, index=index)

ax = df.plot.bar(rot=0, figsize=(10, 7))
```

```python
import datetime

pids = []
times = []

with open("messages", 'r') as f:
    for line in f.readlines():
        line = line.strip()
        try: 
            time, pid = line.split(" ")
            times.append(datetime.datetime.strptime(time, '%H:%M:%S'))
            pids.append(int(pid))
        except: continue
        
        
```

```python
len(times)
```

```python
import matplotlib.pyplot as plt
from matplotlib.dates import date2num
import datetime

# Convert dates to numbers
date_nums = date2num(times)

# Create scatter plot
fig, ax = plt.subplots()
ax.scatter(date_nums, pids, s=1)

# Format x-axis as dates
ax.xaxis_date()
fig.autofmt_xdate()
plt.show()
```
